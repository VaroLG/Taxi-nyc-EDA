El IQR se calcula para cada columna en column_list, y los valores que exceden el umbral superior para cada columna se imputan con el valor de dicho umbral.

```python
def outlier_imputer(column_list, iqr_factor):
  
    for col in column_list:
        # Reassign minimum to zero
        df.loc[df[col] < 0, col] = 0

        # Calculate upper threshold
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        upper_threshold = q3 + (iqr_factor * iqr)
        print(col)
        print('q3:', q3)
        print('upper_threshold:', upper_threshold)

        # Reassign values > threshold to threshold
        df.loc[df[col] > upper_threshold, col] = upper_threshold
        print(df[col].describe())
        print()
```

Definir la hora punta como las franjas de tiempo de 06 a 10 y de 16 a 10 los días entre semana.

```python
# Create 'rush_hour' col
df['rush_hour'] = df['tpep_pickup_datetime'].dt.hour

# If day is Saturday or Sunday, impute 0 in `rush_hour` column
df.loc[df['day'].isin(['saturday', 'sunday']), 'rush_hour'] = 0
```

```python
def rush_hourizer(hour):
    if 6 <= hour['rush_hour'] < 10:
        val = 1
    elif 16 <= hour['rush_hour'] < 20:
        val = 1
    else:
        val = 0
    return val
```

```python
df.loc[(df.day != 'saturday') & (df.day != 'sunday'), 'rush_hour'] = df.apply(rush_hourizer, axis=1)
df.head()
```

Creación de dataset de características y de variable objetivo

```python
X = df2.drop(columns=['fare_amount'])
y = df2[['fare_amount']]
X.head()
```

Para entrenar un modelo con variables categóricas hay que codificarlas como variables dummie, ya que en formato texto no pueden ser introducidas al modelo y como variable categórica numérica alteraría los pesos del modelo.

```python
X['VendorID'] = X['VendorID'].astype(str)
X = pd.get_dummies(X, drop_first=True)
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Estandarizamos las variables de entrenamiento, para que no tenga mas peso una que otra

```python
# Standardize the X variables
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
print('X_train scaled:', X_train_scaled)

X_test_scaled = scaler.transform(X_test)
print('X_test scaled:', X_test_scaled)
```

# Entrenamiento del modelo: Regresión lineal

```python
lr=LinearRegression()
lr.fit(X_train_scaled, y_train)
```

# Evaluación del modelo

Métricas de evaluación de la regresión lineal (R2, MAE "error absoluto medio", MSE " error cuadrático medio", RMSE) con los datos de entrenamiento

```python
r_sq = lr.score(X_train_scaled, y_train)
print('Coefficient of determination:', r_sq)
y_pred_train = lr.predict(X_train_scaled)
print('R^2:', r2_score(y_train, y_pred_train))
print('MAE:', mean_absolute_error(y_train, y_pred_train))
print('MSE:', mean_squared_error(y_train, y_pred_train))
print('RMSE:',np.sqrt(mean_squared_error(y_train, y_pred_train)))
```

Métricas de evaluación de la regresión lineal (R2, MAE "error absoluto medio", MSE " error cuadrático medio", RMSE) con los datos de test

```python
r_sq_test = lr.score(X_test_scaled, y_test)
print('Coefficient of determination:', r_sq_test)
y_pred_test = lr.predict(X_test_scaled)
print('R^2:', r2_score(y_test, y_pred_test))
print('MAE:', mean_absolute_error(y_test,y_pred_test))
print('MSE:', mean_squared_error(y_test, y_pred_test))
print('RMSE:',np.sqrt(mean_squared_error(y_test, y_pred_test)))
```

R2 = 0.868 significa que el modelo explica el 86,8% de la varianza para la variable "fare amount".
